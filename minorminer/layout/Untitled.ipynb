```python
from itertools import product
import networkx as nx

import dwave_networkx as dnx
import minorminer as mm
```

```python
G = nx.complete_bipartite_graph(8,8)
```

```python
G.add_edges_from([e for e in product(range(4),range(4,8))])
```

```python
G.add_edges_from([e for e in product(range(8,12),range(12,16))])
```

```python
T = dnx.chimera_graph(3)
```

```python
emb = mm.find_embedding(G, T)
```

```python
dnx.draw_chimera_embedding(T, emb)
```
